Using the Coin Market Cap API

```
%config IPCompleter.greedy=True
import pandas as pd
import numpy as np

import json
import requests
import datetime
import time
import locale

format = "%Y-%m-%d"
```

```
def topCMC(tokens):
    cmc = json.loads(requests.get("https://api.coinmarketcap.com/v1/ticker/?start=0&limit="+str(tokens)).text)
    new = []
    for i in range(tokens):
      new.append(
          [cmc[i]["rank"],
           cmc[i]["name"],
           cmc[i]["symbol"],
           cmc[i]["price_usd"],
           cmc[i]["24h_volume_usd"],
           cmc[i]["market_cap_usd"],
           cmc[i]["available_supply"],
           cmc[i]["total_supply"],
           cmc[i]["max_supply"],
           cmc[i]["percent_change_1h"],
           cmc[i]["percent_change_24h"],
           cmc[i]["percent_change_7d"],
          ])

    df2 = pd.DataFrame(new, columns = ['rank', 'name', 'symbol', 'price',
                                       '24h_volume', 'market_cap','available_supply',
                                       'total_supply','max_supply', 'percent_change_1h',
                                       'percent_change_24h','percent_change_7d'])
                                  
    df2['rank'] = pd.to_numeric(df2['rank'])
    df2['price'] = pd.to_numeric(df2['price'])
    df2['24h_volume'] = pd.to_numeric(df2['24h_volume'])
    df2['market_cap'] = pd.to_numeric(df2['market_cap'])
    df2['available_supply'] = pd.to_numeric(df2['available_supply'])
    df2['total_supply'] = pd.to_numeric(df2['total_supply'])
    df2['max_supply'] = pd.to_numeric(df2['max_supply'])
    df2['total_supply'] = pd.to_numeric(df2['total_supply'])
    df2['percent_change_1h'] = pd.to_numeric(df2['percent_change_1h'])
    df2['percent_change_24h'] = pd.to_numeric(df2['percent_change_24h'])
    df2['percent_change_7d'] = pd.to_numeric(df2['percent_change_7d'])
    return df2

```

```
df2 = topCMC(1000)
print df2.head()
```

Market Cap under 25,000,000 Positive % change last 7d, but under 300%


```
smallmrkt = df2[(df2.market_cap <= 25000000) & (df2.percent_change_7d >= 0) & (df2.percent_change_7d <= 300)]

smallmrkt.sort_values(by = 'percent_change_7d', ascending=False).head(10)
```

Best performing last 24 hours, over 100,000 in 24h volume and top 250 marketcap rank


```
best24 = df2[ (df2['24h_volume'] > 100000) & (df2['rank'] < 250)]

best24.sort_values(by = 'percent_change_24h', ascending=False).head(10)
```

Best performing last 7 days, and has a max supply, top 500

```
max7 = df2[ (df2['24h_volume'] > 100000) & (df2['rank'] < 500)].dropna(subset = ['max_supply'])

max7.sort_values(by = 'percent_change_7d', ascending=False).head(10)
```
