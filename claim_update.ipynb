```python
df = pd.read_csv('~/Downloads/test_data.csv')

#df['external_id'] = df['external_id'].astype(str)   
#df['old_id'] = df['old_id'].astype(str)      
#df['old_id'] = df['old_id'].str.replace('.0', '')

df.head()

```

```python
list_old = list(map(str, df.old_id.tolist()))
list_new = list(map(str, df.external_id.tolist()))

df['new_claim'] = df.claim.replace(to_replace=['Claim ID: : ' + e for e in list_old], value=['Claim ID: ' + e for e in list_new], regex=True)
df['new_description'] = df.description.replace(to_replace=['\* ' + e + ' \*' for e in list_old], value=['* ' + e + ' *' for e in list_new], regex=True)

df.head()
```
